```python
import pandas as pd
```

```python
g1 = pd.read_csv('data/groups1.csv')
g2 = pd.read_csv('data/groups2.csv')
cat = pd.read_csv('data/categories.csv')
city = pd.read_csv('data/cities.csv', dtype={'zip':'string'})
```

```python
groups = pd.concat([g1, g2], ignore_index=True)
groups
```

```python
groups.merge(right=cat, how='left', on='category_id') # left inner join
```

```python
groups.merge(right=cat, how='inner', on='category_id') # inner join (equi join )
```

```python
groups.merge(right=cat, how='outer', on='category_id') # outter join
```

```python
# when the columns haven't the same name use below way:
groups.merge(right=city, how='outer', left_on='city_id', right_on='id')
```

```python
outer = groups.merge(right = city, how = 'outer', left_on = 'city_id', right_on = 'id', indicator = True)
outer
```
