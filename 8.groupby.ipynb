```python
import pandas as pd
```

```python
food_data = {
    'item':['banana', 'cucumber', 'ornage', 'tomato', 'watermelon'],
    'type':['fruit', 'vegetable', 'fruit', 'vegetable', 'fruit'],
    'price':[0.99, 1.25, 0.25, 0.33, 3.00]
}

super_market = pd.DataFrame(data=food_data)
super_market
```

```python
g = super_market.groupby(by='type')
```

```python
g.sum(numeric_only=True)
```

```python
g.get_group('fruit').sum()
```
