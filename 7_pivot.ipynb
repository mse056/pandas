```python
import pandas as pd
```

```python
ds = pd.read_csv('data/sales_by_employee.csv', parse_dates = ['Date'])
ds
```

```python
ds.pivot_table(index = 'Date', values =['Expenses', 'Revenue']) # default aggfunc = 'mean'
```

```python
ds.pivot_table(index = 'Date', aggfunc = 'sum', values =['Expenses', 'Revenue'])
```

```python
ds.pivot_table(index = 'Date', aggfunc = 'sum', values =['Expenses', 'Revenue'], columns='Name')
```

```python
ds.pivot_table(index = 'Date', aggfunc = 'sum', values =['Expenses', 'Revenue'], columns = 'Name', margins = True)
```

```python
ds.pivot_table(index = 'Date', aggfunc = 'sum', values =['Expenses', 'Revenue'], columns = 'Name', margins = True,
                margins_name = 'Total')
```

```python
ds.pivot_table(index = 'Date', aggfunc = 'sum', values =['Expenses', 'Revenue'], columns = 'Name',margins = True,
                margins_name = 'Total', fill_value = 0)
```

```python
ds.pivot_table(index = 'Date', columns = 'Name', values =['Expenses', 'Revenue'], aggfunc = ['sum','mean'], fill_value = 0)
```

```python
ds.pivot_table(index = 'Date', columns = 'Name', values =['Expenses', 'Revenue'], 
               aggfunc = {'Expenses':'sum', 'Revenue':'mean'},fill_value = 0)
```

```python
ds.pivot_table(index = ['Date','Name'], values ='Revenue')
```

```python
new = ds.pivot_table(index = 'Name', columns = 'Date', values ='Revenue', aggfunc = 'sum')
new
```

```python
new.stack()
```

```python
new = ds.pivot_table(index = ['Customer', 'Name'], values ='Revenue', aggfunc = 'sum')
new
```

```python
new.unstack()
```
